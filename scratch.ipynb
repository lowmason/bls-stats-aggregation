```python
from bls_stats import download_qcew_bulk, download_jolts
from bls_stats.qcew import map_bulk_to_ces
from bls_stats.jolts import map_jolts_to_ces

import polars as pl
```

```python
ces_bulk_path = download_qcew_bulk(start_year=2024, end_year=2024)
```

```python
ces = (
    map_bulk_to_ces(
        ces_bulk_path
    )
    .sort(
        'ref_date',
        'geographic_type', 'geographic_code',
        'industry_type', 'industry_code'
    )
)
```

```python
print('Industry codes:', ces['industry_code'].unique().sort().to_list())
print()
nat = ces.filter(ces['geographic_type'] == 'national').sort('industry_code')
for row in nat.unique(subset=['industry_code', 'industry_type']).sort('industry_code').iter_rows(named=True):
    print(f"  {row['industry_code']:>3s}  {row['industry_type']}")
```

```python
jolts_path = download_jolts()
```

```python
jolts = (
    map_jolts_to_ces(jolts_path)
)
jolts.filter(pl.col('geographic_code').eq('56'), pl.col('industry_code').eq('10'))
```

```python
print('Industry codes:', jolts['industry_code'].unique().sort().to_list())
print('Data elements:', jolts['data_element'].unique().to_list())
print('Rate/level:', jolts['rate_or_level'].unique().to_list())
print()
for row in jolts.unique(subset=['industry_code', 'industry_type']).sort('industry_code').iter_rows(named=True):
    print(f"  {row['industry_code']:>3s}  {row['industry_type']}")
```
